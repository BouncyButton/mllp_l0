```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.transforms import Bbox

import math

import wandb
api = wandb.Api()
```

```python
def plot_and_save_f1_error(plot_dict, file_name):
    fig, ax = plt.subplots(figsize=(8, 5.5))
    # plt.figure()
    # fig.tight_layout(pad=2)

    offsets = np.linspace(-0.05, 0.05, len(plot_dict))

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])

        ax.bar(x + offset, y, 0.015)
        ax.legend(plot_dict.keys(), loc="upper left", ncol=2, fontsize=10)
        

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])
        sd = np.array(plot_dict[key]["SD"])
        err = (sd / math.sqrt(5))

        ax.errorbar(x + offset, y, yerr=err, label=key, capsize=5, linestyle='None')


    
    plt.xlabel("Random Binarization Rate", fontsize="20")
    plt.ylabel("F1 Score", fontsize="20")

    plt.ylim(0, 0.8)
    plt.yticks(np.arange (0, 1.0, 0.2), fontsize=14)
    plt.xticks([0.0, 0.5, 0.9], fontsize=14)

    for ytick in np.arange (0, 1.0, 0.2):
        ax.axhline(y=ytick, color = 'grey', linewidth=0.2)

    chartBox = ax.get_position()
    # Manually adjust the position of the plot within the figure
    # The parameters are [left, bottom, width, height] in relative figure coordinates
    # Decrease the bottom parameter to add space below the plot
    ax.set_position(Bbox([[0.09, 0.20], [0.97, 0.9]]))  # Adjust this as needed


    fig.savefig(file_name)
    plt.show()

def plot_and_save_rules_size(plot_dict, file_name):
    fig, ax = plt.subplots(figsize=(8, 5.5))
    # plt.figure()
    # fig.tight_layout(pad=2)

    offsets = np.linspace(-0.05, 0.05, len(plot_dict))

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])

        ax.bar(x + offset, y, 0.015)
        ax.legend(plot_dict.keys(), loc="upper left", ncol=2, fontsize=10)
        

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])
        sd = np.array(plot_dict[key]["SD"])
        err = (sd / math.sqrt(5))

        ax.errorbar(x + offset, y, yerr=err, label=key, capsize=5, linestyle='None')


    
    plt.xlabel("Random Binarization Rate", fontsize="20")
    plt.ylabel("Total Length of Rules", fontsize="20")

    plt.ylim(0, 10000)
    plt.yticks(np.arange (0, 11000, 1000), fontsize=14)
    plt.xticks([0.0, 0.5, 0.9], fontsize=14)

    for ytick in np.arange (0, 11000, 1000):
        ax.axhline(y=ytick, color = 'grey', linewidth=0.2)

    chartBox = ax.get_position()
    # Manually adjust the position of the plot within the figure
    # The parameters are [left, bottom, width, height] in relative figure coordinates
    # Decrease the bottom parameter to add space below the plot
    ax.set_position(Bbox([[0.15, 0.20], [0.97, 0.9]]))  # Adjust this as needed


    fig.savefig(file_name)
    plt.show()


def plot_and_save_active_weights(plot_dict, file_name):
    fig, ax = plt.subplots(figsize=(8, 5.5))
    # plt.figure()
    # fig.tight_layout(pad=2)

    offsets = np.linspace(-0.05, 0.05, len(plot_dict))

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])

        ax.bar(x + offset, y, 0.015)
        ax.legend(plot_dict.keys(), loc="upper left", ncol=2, fontsize=10)
        

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])
        sd = np.array(plot_dict[key]["SD"])
        err = (sd / math.sqrt(5))

        ax.errorbar(x + offset, y, yerr=err, label=key, capsize=5, linestyle='None')


    
    plt.xlabel("Random Binarization Rate", fontsize="20")
    plt.ylabel("Total Active Weights", fontsize="20")

    plt.ylim(0, 140000)
    plt.yticks(np.arange (0, 160000, 20000), fontsize=14)
    plt.xticks([0.0, 0.5, 0.9], fontsize=14)

    for ytick in np.arange (0, 160000, 20000):
        ax.axhline(y=ytick, color = 'grey', linewidth=0.2)

    chartBox = ax.get_position()
    # Manually adjust the position of the plot within the figure
    # The parameters are [left, bottom, width, height] in relative figure coordinates
    # Decrease the bottom parameter to add space below the plot
    ax.set_position(Bbox([[0.15, 0.20], [0.97, 0.9]]))  # Adjust this as needed


    fig.savefig(file_name)
    plt.show()

def plot_and_save_execution_time(plot_dict, file_name):
    fig, ax = plt.subplots(figsize=(8, 5.5))
    # plt.figure()
    # fig.tight_layout(pad=2)

    offsets = np.linspace(-0.05, 0.05, len(plot_dict))

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])

        ax.bar(x + offset, y, 0.015)
        ax.legend(plot_dict.keys(), loc="upper left", ncol=2, fontsize=10)

    for key, offset in zip(plot_dict.keys(), offsets):
        x = np.array(plot_dict[key]["P"])
        y = np.array(plot_dict[key]["Mean"])
        sd = np.array(plot_dict[key]["SD"])
        err = (sd / math.sqrt(5))

        ax.errorbar(x + offset, y, yerr=err, label=key, capsize=5, linestyle='None')
        
    plt.xlabel("Random Binarization Rate", fontsize="20")
    plt.ylabel("Execution Time", fontsize="20")

    plt.ylim(0, 5000)
    plt.yticks(np.arange (0, 5500, 500), fontsize=14)
    plt.xticks([0.0, 0.5, 0.9], fontsize=14)

    for ytick in np.arange (0, 5500, 500):
        ax.axhline(y=ytick, color = 'grey', linewidth=0.2)

    chartBox = ax.get_position()
    # Manually adjust the position of the plot within the figure
    # The parameters are [left, bottom, width, height] in relative figure coordinates
    # Decrease the bottom parameter to add space below the plot
    ax.set_position(Bbox([[0.15, 0.20], [0.97, 0.9]]))  # Adjust this as needed


    fig.savefig(file_name)
    plt.show()

def process_runs_f1_wandb(project):
    runs = api.runs(f"mllp_l0/{project}")

    results = []

    for run in runs:

        result = {}

        result["Model"] = "MLLP"
        result["P"] = run.config["random_binarization_rate"]
        result["Structure"] = run.config["structure"]
        result["Mean"] = run.summary["test_f1_score_kf_avg"]
        result["SD"] = run.summary["test_f1_score_kf_sd"]
        results.append(result)

        result = {}

        result["Model"] = "CRS"
        result["P"] = run.config["random_binarization_rate"]
        result["Structure"] = run.config["structure"]
        result["Mean"] = run.summary["test_f1_score_b_kf_avg"]
        result["SD"] = run.summary["test_f1_score_b_kf_sd"]
        results.append(result)

    df = pd.DataFrame(results)

    df_crs = df[df["Model"] == "CRS"][["P", "Structure", "Mean", "SD"]]
    df_mllp = df[df["Model"] == "MLLP"][["P", "Structure", "Mean", "SD"]]

    plot_dict_crs = df_crs[df_crs["Structure"].isin(['64', '128', '256', '64_64_64', '128_128_128', '256_256_256'])].groupby("Structure").agg({"Mean": list, "SD": list, "P": list}).sort_values(by="Structure", key=lambda x: x.map({'32': 0,
    '64': 1,
    '128': 2,
    '256': 3,
    '32_32_32': 4,
    '64_64_64': 5,
    '128_128_128': 6,
    '256_256_256': 7})).to_dict("index")

    plot_dict_mllp = df_mllp[df_mllp["Structure"].isin(['64', '128', '256', '64_64_64', '128_128_128', '256_256_256'])].groupby("Structure").agg({"Mean": list, "SD": list, "P": list}).sort_values(by="Structure", key=lambda x: x.map({'32': 0,
    '64': 1,
    '128': 2,
    '256': 3,
    '32_32_32': 4,
    '64_64_64': 5,
    '128_128_128': 6,
    '256_256_256': 7})).to_dict("index")

    return plot_dict_crs, plot_dict_mllp

def process_runs_rules_size_wandb(project):
    runs = api.runs(f"mllp_l0/{project}")

    results = []

    for run in runs:

        result = {}

        result["Model"] = "CRS"
        result["P"] = run.config["random_binarization_rate"]
        result["Structure"] = run.config["structure"]
        result["Mean"] = run.summary["total_rules_sizes_kf_avg"]
        result["SD"] = run.summary["total_rules_sizes_kf_sd"]
        results.append(result)

    df = pd.DataFrame(results)

    df_crs = df[df["Model"] == "CRS"][["P", "Structure", "Mean", "SD"]]

    plot_dict_crs = df_crs[df_crs["Structure"].isin(['64', '128', '256', '64_64_64', '128_128_128', '256_256_256'])].groupby("Structure").agg({"Mean": list, "SD": list, "P": list}).sort_values(by="Structure", key=lambda x: x.map({'32': 0,
    '64': 1,
    '128': 2,
    '256': 3,
    '32_32_32': 4,
    '64_64_64': 5,
    '128_128_128': 6,
    '256_256_256': 7})).to_dict("index")

    return plot_dict_crs

def process_runs_active_weights_wandb(project, active_weights_variable):
    runs = api.runs(f"mllp_l0/{project}")

    results = []

    for run in runs:

        result = {}

        result["Model"] = "CRS"
        result["P"] = run.config["random_binarization_rate"]
        result["Structure"] = run.config["structure"]
        result["Mean"] = run.summary[f"{active_weights_variable}_avg"]
        result["SD"] = run.summary[f"{active_weights_variable}_sd"]
        results.append(result)

    df = pd.DataFrame(results)

    df_mllp = df[df["Model"] == "CRS"][["P", "Structure", "Mean", "SD"]]

    plot_dict_mllp = df_mllp[df_mllp["Structure"].isin(['64', '128', '256', '64_64_64', '128_128_128', '256_256_256'])].groupby("Structure").agg({"Mean": list, "SD": list, "P": list}).sort_values(by="Structure", key=lambda x: x.map({'32': 0,
    '64': 1,
    '128': 2,
    '256': 3,
    '32_32_32': 4,
    '64_64_64': 5,
    '128_128_128': 6,
    '256_256_256': 7})).to_dict("index")

    return plot_dict_mllp

def process_runs_execution_time_wandb(project):
    runs = api.runs(f"mllp_l0/{project}")

    results = []

    for run in runs:

        result = {}

        result["P"] = run.config["random_binarization_rate"]
        result["Structure"] = run.config["structure"]
        result["Mean"] = run.summary["run_time_kf_avg"]
        result["SD"] = run.summary["run_time_kf_sd"]
        results.append(result)

    df = pd.DataFrame(results)

    df = df[["P", "Structure", "Mean", "SD"]]

    plot_dict = df[df["Structure"].isin(['64', '128', '256', '64_64_64', '128_128_128', '256_256_256'])].groupby("Structure").agg({"Mean": list, "SD": list, "P": list}).sort_values(by="Structure", key=lambda x: x.map({'32': 0,
    '64': 1,
    '128': 2,
    '256': 3,
    '32_32_32': 4,
    '64_64_64': 5,
    '128_128_128': 6,
    '256_256_256': 7})).to_dict("index")

    return plot_dict


```
